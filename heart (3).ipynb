# sepate category columns & numeric columns

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x.columns
```

```python
cat_col=x.select_dtypes(include="O")
num_col=x.select_dtypes(include=["int","float"])
```

```python
cat_col.columns
```

```python
num_col.columns
```

# Outliers

```python
for i in num_col:
    sns.boxplot(x=data[i])
    plt.show()
```

# skewness

```python
num_col.skew()
```

```python
b=[]
for i in num_col:
    a=x[i].skew()
    if a>0.5 or a<-0.5 :
        b=b+[i]
print(b)
```

```python
from sklearn .preprocessing import PowerTransformer
trans=PowerTransformer(method="yeo-johnson")
x[b]=trans.fit_transform(x[b])
```

```python
num_col.skew()
```

# Pipeline with scaling encoding and smote

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_col_pip=Pipeline(steps=[("one",OneHotEncoder())])
num_col_pip=Pipeline(steps=[("scaler",StandardScaler())])
```

# column transform

```python
from sklearn.compose import ColumnTransformer
```

```python
cat_cols=cat_col.columns
num_cols=num_col.columns
```

```python
trans=ColumnTransformer(transformers=[("cat",cat_col_pip,cat_cols),("num",num_col_pip,num_cols)])
```

```python
trans
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=6)
```

# model Implementaion

## Logistic Regression

```python
from imblearn.pipeline import Pipeline
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_log=Pipeline(steps=[("preprocessing",trans),("smot",smote),("model",LogisticRegression())])
```

```python
model_log
```

```python
model_log.fit(x_train,y_train)
```

```python

from sklearn.metrics import accuracy_score , f1_score
```

```python
pre_log=model_log.predict(x_test)
```

```python
accuracy_score(pre_log,y_test)
```

```python
f1_score(pre_log,y_test)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dec=Pipeline(steps=[("preprocessing",trans),("smot",smote),("model",DecisionTreeClassifier(random_state=42))])
```

```python
model_dec.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score , f1_score
pre_dec=model_dec.predict(x_test)
accuracy_score(pre_dec,y_test)
```

```python
f1_score(pre_dec,y_test)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_ran=Pipeline(steps=[("preprocessing",trans),("smot",smote),("model",RandomForestClassifier(random_state=42))])
```

```python
model_ran.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score , f1_score
pre_ran=model_ran.predict(x_test)
accuracy_score(pre_ran,y_test)
```

```python
f1_score(pre_ran,y_test)
```

```python
models = {
    "Logistic Regression": accuracy_score(y_test, pre_log),
    "Decision Tree": accuracy_score(y_test, pre_dec),
    "Random Forest": accuracy_score(y_test, pre_ran)
    
}

model_df = pd.DataFrame(list(models.items()), columns=['Model', 'Accuracy'])
model_df.sort_values(by='Accuracy', ascending=False)
```
